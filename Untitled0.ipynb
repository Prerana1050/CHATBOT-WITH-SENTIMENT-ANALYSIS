```python
# sentiment.py logic inside Colab

from nltk.sentiment.vader import SentimentIntensityAnalyzer

class SentimentAnalyzer:
    def __init__(self):
        self.sia = SentimentIntensityAnalyzer()

    def analyze_statement(self, text):
        scores = self.sia.polarity_scores(text)
        compound = scores['compound']
        if compound >= 0.05:
            label = "Positive"
        elif compound <= -0.05:
            label = "Negative"
        else:
            label = "Neutral"
        return {**scores, "label": label}

    def analyze_conversation(self, messages):
        results = [self.analyze_statement(msg) for msg in messages]
        compounds = [r['compound'] for r in results]
        mean_c = sum(compounds) / len(compounds) if compounds else 0.0

        if mean_c >= 0.05:
            overall = "Positive"
        elif mean_c <= -0.05:
            overall = "Negative"
        else:
            overall = "Neutral"

        return overall, results
```

```python
# chatbot.py logic inside Colab

sentiment = SentimentAnalyzer()
conversation = []

print("Chatbot: Hello! Let's chat. Type 'done' to finish.\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "done":
        break

    conversation.append(user_input)

    # Very simple response
    reply = "Interesting! Tell me more..."
    print(f"Chatbot: {reply}")

# After conversation ends
overall, details = sentiment.analyze_conversation(conversation)

print("\n--- Sentiment Analysis ---")
for i, d in enumerate(details, start=1):
    print(f"{i}. {conversation[i-1]} => {d['label']} ({d['compound']:.3f})")

print(f"\nOverall Conversation Sentiment: {overall}")
```
